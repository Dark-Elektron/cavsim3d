# Rectangular Waveguide Analysis

This notebook demonstrates a complete rectangular waveguide analysis workflow:

1. Geometry creation
2. Frequency-domain analysis
3. Model order reduction
4. Comparison with analytical solution

```python
from utils.visualization import *
from geometry.importers import STEPImporter
from solvers.frequency_domain import FrequencyDomainSolver
from rom.reduction import ModelOrderReduction
from analytical.rectangular_waveguide import RWGAnalytical
from ngsolve.webgui import Draw # must import Draw, otherwise may run into problems showing mesh

%matplotlib inline
plt.rcParams['figure.dpi'] = 100
```

## 1. Define Geometry

Create a rectangular waveguide with specified dimensions and mesh parameters.

```python
# === Complete workflow example ===

from geometry.importers import STEPImporter
from solvers import FrequencyDomainSolver, ConcatenatedSystem
from rom.reduction import ModelOrderReduction
import matplotlib.pyplot as plt

# 1. Load and prepare geometry
geo = STEPImporter(r"./rectangular_waveguide.step", auto_build=False)
geo.add_splitting_plane_at_z(0.05)
geo.add_splitting_plane_at_z(0.10)
geo.split()
# geo.show_split_preview()
geo.build()
geo.name_solids()
# geo.finalize(maxh=0.02)
geo.generate_mesh(maxh=0.04) # after naming solids, must generate mesh but avoid rebuilding

# 2. Full-order solve
fds = FrequencyDomainSolver(geo, order=3)
fds.assemble_matrices(nmodes=1)
fds.solve(1, 3, 30, store_snapshots=True)

# 3. Get FOM eigenvalues
eigs_fom = fds.get_eigenvalues()
freqs_fom = fds.get_resonant_frequencies()
print(f"FOM resonances: {freqs_fom[:5]/1e9} GHz")

# 4. Model order reduction
rom = ModelOrderReduction(fds)
rom.reduce(tol=1e-6)

# 5. Get ROM eigenvalues (per domain)
eigs_rom = rom.get_eigenvalues()
freqs_rom = rom.get_resonant_frequencies()
print(f"ROM resonances: {freqs_rom[:5]/1e9} GHz")

# 6. Solve ROM (auto-concatenates for multi-domain)
rom.solve(1, 3, 100)

# 7. Compare FOM vs ROM
rom.compare_with(fds, labels=['ROM', 'FOM'])

# 8. Access concatenated system if needed
if rom.n_domains > 1:
    concat = rom.concatenate()
    concat.print_info()
    
    # Solve concatenated system
    concat.solve(1, 3, 100)
    
    # Further reduce the concatenated system
    concat_reduced = concat.reduce(tol=1e-8)
    concat_reduced.solve(1, 3, 100)
    
    # Compare all levels
    print("\n=== DOF Comparison ===")
    print(f"FOM DOFs: {sum(fds.M[d].shape[0] for d in fds.domains)}")
    print(f"ROM DOFs (per domain): {rom.total_reduced_dofs}")
    print(f"Concat DOFs: {concat.A_coupled.shape[0]}")
    print(f"Reduced Concat DOFs: {concat_reduced.A_coupled.shape[0]}")
    
    # Proper S-parameter comparison (mapping external ports correctly)
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    freq_ghz_fom = fds.frequencies / 1e9
    freq_ghz_rom = rom.frequencies / 1e9
    freq_ghz_concat = concat.frequencies / 1e9
    freq_ghz_reduced = concat_reduced.frequencies / 1e9
    
    # S11 (reflection at input)
    ax = axes[0]
    ax.plot(freq_ghz_fom, fds.get_s_db(1, 1), 'b-', label='FOM S11', linewidth=2)
    ax.plot(freq_ghz_rom, rom.get_s_db(1, 1), 'r--', label='ROM S11', linewidth=1.5)
    ax.plot(freq_ghz_concat, concat.get_s_db(1, 1), 'g:', label='Concat S11', linewidth=1.5)
    ax.plot(freq_ghz_reduced, concat_reduced.get_s_db(1, 1), 'm-.', label='ReducedConcat S11', linewidth=1.5)
    ax.set_xlabel('Frequency (GHz)')
    ax.set_ylabel('|S11| (dB)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_title('S11 Comparison (Input Reflection)')
    
    # S21 (transmission) - NOTE: FOM S14 = ROM S12
    ax = axes[1]
    n_ports_fom = fds.n_ports
    ax.plot(freq_ghz_fom, fds.get_s_db(1, n_ports_fom), 'b-', label=f'FOM S1{n_ports_fom}', linewidth=2)
    ax.plot(freq_ghz_rom, rom.get_s_db(1, 2), 'r--', label='ROM S12', linewidth=1.5)
    ax.plot(freq_ghz_concat, concat.get_s_db(1, 2), 'g:', label='Concat S12', linewidth=1.5)
    ax.plot(freq_ghz_reduced, concat_reduced.get_s_db(1, 2), 'm-.', label='ReducedConcat S12', linewidth=1.5)
    ax.set_xlabel('Frequency (GHz)')
    ax.set_ylabel('|S21| (dB)')
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.set_title('S21 Comparison (Transmission)')
    
    plt.tight_layout()
    plt.show()

# 9. Per-domain comparison
rom_per_domain = rom.solve_per_domain()
for domain in rom.domains:
    fds_domain = fds.get_domain_results(domain)
    rom_domain = rom_per_domain[domain]
    print(f"\nDomain {domain}:")
    print(f"  FDS Z[0,0] at f=5GHz: {fds_domain['Z'][20,0,0]}")
    print(f"  ROM Z[0,0] at f=5GHz: {rom_domain['Z'][20,0,0]}")
```

## New Result-Object API

The new API provides a navigable object graph:
- `fds.fom` — global FOM result
- `fds.fom.rom` — ROM of the global FOM
- `fds.foms` — per-domain FOMs (multi-solid only)
- `fds.foms.roms.concat.rom` — full ROM concatenation chain

Each result object has `.plot_s()`, `.plot_z()`, `.plot_eigenvalues()` methods.

```python
# === New API: result-object navigation ===

# FOM result — same data as fds, but as a standalone result object
fom_result = fds.fom
print(fom_result)

# Plot S-parameters directly from the FOM result
fig, ax = fom_result.plot_s(params=['1(1)1(1)', '1(1)2(1)'], label='FOM')

# ROM result — navigate from FOM
rom_result = fds.fom.rom
print(rom_result)

# Overlay ROM onto the same axes
fig, ax = rom_result.plot_s(params=['1(1)1(1)', '1(1)2(1)'], ax=ax, label='ROM')
plt.show()
```

## 2. Analytical Solution

Compute the analytical Z-parameters for comparison with numerical results.

```python
# Waveguide parameters
a = 100e-3  # Width: 100 mm
L = 200e-3  # Length: 200 mm

analytical = RWGAnalytical(a=a, L=L)
frequencies = np.linspace(1.0, 3.0, 100) * 1e9

Z_analytical = analytical.z_parameters(frequencies)
S_analytical = analytical.s_parameters(frequencies, Z0_ref='ZTE')

print(f"Computed analytical Z and S parameters at {len(frequencies)} points")
print(f"Cutoff frequency: {analytical.fc/1e9:.4f} GHz")
```

## 3. Full-Order Numerical Solution

Run a full-order finite element frequency sweep.

```python
# Create solver with wave impedance enabled
solver = FrequencyDomainSolver(rwg, order=3)
solver.assemble_matrices()

# Print port info
solver.print_port_info()
```

```python
solver.port_solver.port_orientation_factors
```

```python
# Solve with S-parameter computation
results = solver.solve(fmin=1.0, fmax=3.0, nsamples=30,
                       store_snapshots=True, compute_s_params=True)
freq = results['frequencies']
Z_full = results['Z']
S_full = results['S']

print(f"Solved at {len(freq)} frequency points")
print(f"Z-parameter keys: {[k for k in Z_full.keys() if k != 'frequencies']}")
print(f"S-parameter keys: {[k for k in S_full.keys() if k != 'frequencies']}")
```

```python
solver.plot_port_mode('port4', mode=0, component='abs')
```

```python
# Plot with clipping plane
solver.plot_field(
    freq_idx=25,  # Middle frequency
    component='abs',
    domain='cell_3'
)
```

```python
# Plot field along waveguide axis
fig, ax = solver.plot_field_along_line(
    freq_idx=25,
    start=(a/2, a/2, 0),
    end=(a/2, a/2, L),
    component='abs',
    domain='cell_1'
)
plt.show()
```

## 4. Model Order Reduction

Build a reduced-order model using Proper Orthogonal Decomposition (POD).

```python
# Build reduced-order model
rom = ModelOrderReduction(solver)
rom.reduce(tol=1e-6, max_rank=30)
```

```python

# Solve with cascade (fast)
results = rom.solve(method='cascade')

# # Or coupled (more accurate)
# results = rom.solve(method='coupled')
rom.plot_singular_values()
Z_reduced = results['Z']
```

## 5. Comparison and Plotting

Compare Z-parameters from analytical, full FEM, and ROM solutions.

```python
plotter = ZParameterPlotter(figsize=(12, 10))
plotter.setup_figure()
plotter.plot_comparison(Z_analytical, Z_full, Z_reduced)
plt.suptitle('Z-Parameter Comparison', fontsize=14)
plt.show()
```

```python
plotter = SParameterPlotter(figsize=(12, 10))
plotter.setup_figure()
plotter.plot_comparison(S_analytical, S_full, Z_reduced)
plt.suptitle('S-Parameter Comparison', fontsize=14)
plt.show()
```

```python
ports = list(solver.port_modes.keys())
Zw_numerical = np.array([solver.port_solver.get_wave_impedance(ports[0], 0, f) for f in frequencies])
Zw_analytical = analytical.wave_impedance(frequencies)

# Plot
fig, ax = WaveImpedancePlotter.quick_plot(
    frequencies, Zw_analytical, Zw_numerical,
    cutoff_freq=analytical.fc,
    title='TE₁₀ Wave Impedance'
)
plt.show()
```

```python
# Singular value decay plot
fig2, ax2 = plt.subplots(figsize=(8, 5))

ConvergencePlotter.plot_singular_values(
    rom._singular_values,
    truncation_rank=rom.r,
    ax=ax2
)

plt.title('POD Singular Value Decay')
plt.tight_layout()
plt.show()
```

## 6. Error Analysis

Quantify the error between numerical solutions and the analytical reference.

```python
# ROM error
Z_ana_rom = analytical.z_parameters(Z_reduced['frequencies'])
Z11_rom = Z_reduced['Z'][:, 0, 0]
Z11_ana_rom = Z_ana_rom['Z11']

rel_error_rom = np.abs(Z11_rom - Z11_ana_rom) / np.abs(Z11_ana_rom)

print("ROM vs Analytical:")
print(f"   Mean relative error: {np.mean(rel_error_rom) * 100:.2f}%")
print(f"   Max relative error:  {np.max(rel_error_rom) * 100:.2f}%")
```

## 7. Eigenfrequency Comparison

Compare eigenfrequencies from the ROM with analytical values.

```python
# Get ROM eigenfrequencies
eigs_rom = rom.get_eigenvalues()
freqs_rom = np.sqrt(np.abs(eigs_rom)) / (2 * np.pi)
freqs_rom_ghz = freqs_rom / 1e9

print("ROM eigenfrequencies (GHz):")
for i, f in enumerate(sorted(freqs_rom_ghz)[:10]):
    print(f"   Mode {i + 1}: {f:.4f} GHz")
```
